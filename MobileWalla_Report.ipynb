```python
df1.columns = ['ifa', 'city','day_part','geohash','user_age','device_name','device_os','device_manufacturer','device_platform','user_gender']
```

```python
df01 = df1[pd.notnull(df1['user_gender'])]
```

```python
df01 = df01[df01.user_age != 0]
```

```python
df01 = df01[df01.device_name != "NIL"]
```

```python
df01 = df01[df01.device_platform != "NIL"]
```

```python
df01 = df01[df01.device_os != "NIL"]
```

```python
df01 = df01[df01.city != "NIL"]
```

```python
list01 = df01.user_age.unique()
```

```python
aDict = {}
for i in list01:
    y = (df01["user_age"]==i).sum()
    aDict[i] = y
```

```python
sorted_aDict = sorted(aDict.items(), key=operator.itemgetter(1))
```

```python
sorted_aDict.reverse()
```

```python
aDict = collections.OrderedDict(sorted_aDict)
```

# User Age Visualized

```python
plt.bar(range(len(aDict)), aDict.values())
plt.xticks(range(len(aDict)), list(aDict.keys()))
plt.gcf().set_size_inches((20, 20)) 
plt.xticks(rotation=70)
plt.show()
```
